```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df_eco = pd.read_csv("Economy_Productivity_SD_India.csv")
df_gdp = pd.read_csv("GDP_Sector_Income_2019_2023_30Cities.csv")
```

```python
df = pd.merge(left=df_eco,right=df_gdp,how='inner')
```

```python
plt.figure(figsize=(12,12))
correlation_matrix = df.drop(columns=['City', 'Year']).corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.savefig("Correlation.png")
plt.show()
```

```python
for i in df.drop(columns='City').columns:
    sns.histplot(df[i], kde=True)
    plt.savefig("Histplot.png")
    plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 8))
for column in df.columns[2:]:
    plt.plot(df['Year'], df[column], marker='o', label=column)
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Yearly Trend Analysis')
plt.legend()
plt.grid(True)
plt.savefig("Yearly Trend Analysis.png")
plt.show()
```

```python
sns.pairplot(df.drop(columns=['City']), diag_kind='kde')
plt.suptitle('Pair Plot Analysis', y=1.02)
plt.savefig("Pair Plot Analysis.png")
plt.show()
```

```python
numeric_columns = df.drop(columns=['City', 'Year']).columns
for column in numeric_columns:
    sns.boxplot(df[column])
    plt.title(f'Box Plot for {column}')
    plt.xlabel(column)
    plt.savefig("Box plot.png")
    plt.show()
```

```python
for column in numeric_columns:
    sns.scatterplot(x=df['Year'], y=df[column])
    plt.title(f'Scatter Plot of {column} vs Year')
    plt.xlabel('Year')
    plt.ylabel(column)
    plt.savefig("Scatter plot.png")
    plt.show()
```

```python
variables = ['R&D Expenditure (% of GDP)', 'GDP (in billion $)', 'Unemployment Rate (%)']
df[variables].plot(figsize=(14, 8))
plt.title('Time Series Analysis of Selected Variables')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.savefig("Time Series Analysis of Selected Variables.png")
plt.show()
```
